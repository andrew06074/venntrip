```python
import pandas as pd

df = pd.read_csv('testing.csv')
df
```

```python
df['Stripped links'] = df['Google link'].str.lstrip("https://www.google.com/maps/place/")

new_cols = df['Stripped links'].str.split("@",n=1,expand=True)

df['trash'] = new_cols[0]
df['keep'] = new_cols[1]

df.drop(columns=['Stripped links'],inplace=True)
df['keep']

new_cols = df['keep'].str.split(",",n=1,expand=True)
df['lat'] = new_cols[0]
df['keep_for_lon'] = new_cols[1]

new_cols = df['keep_for_lon'].str.split(",",n=1,expand=True)
df['lon'] = new_cols[0]

df.drop(columns=['keep_for_lon'],inplace=True)
df.drop(columns=['trash'],inplace=True)
df.drop(columns=['keep'],inplace=True)

df
```
